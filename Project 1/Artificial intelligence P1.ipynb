```python
import datetime
```

```python
Ai_assistant_Name = "KHAN"
```

```python
responses = {
    "hello": "Hey! Great to see you here. What can I do for you today?",
    "hi": "Hi! I'm KHAN, your personal AI buddy. Ask me anything!",
    "how are you": "Honestly? Running at full power and ready to help you! What about you?",
    "what is ai": "AI stands for Artificial Intelligence — it's the science of making machines think, learn, and make decisions like humans.",
    "what is machine learning": "Machine Learning is a branch of AI where computers learn from data instead of being explicitly programmed. Pretty cool, right?",
    "what is deep learning": "Deep Learning is a type of Machine Learning that uses neural networks with many layers to understand complex patterns like images and speech.",
    "bye": "Goodbye! Have a great day.",
    "help": "I can answer questions about AI. Try asking 'what is ai'!",
    "what is decodelabs": "DecodeLabs is a tech training organization that helps students and interns build real-world AI and software projects from scratch.",
    "who made you": "I was built by an AI Engineer in training at DecodeLabs!",
}
```

```python
def get_clean_input(username):
    raw = input(f"{username}: ")
    return raw.lower().strip()
```

```python
def get_response(user_input):
    for key in responses:
        if key in user_input:
            return responses[key]
    return "I do not understand. Could you rephrase?"
```

```python
def main():
    turns = 0

    username = input("Enter your name: ").strip()
    if not username:
        username = "User"

    print(f"{Ai_assistant_Name}: Nice to meet you, {username}! I'm {Ai_assistant_Name}, your AI assistant. Type 'exit' to quit.")

    while True:
        user_input = get_clean_input(username)

        if user_input == "exit":
            print(f"{Ai_assistant_Name}: Goodbye, {username}! We had {turns} exchanges.")
            print(f"Session ended at {datetime.datetime.now().strftime('%H:%M:%S')}")
            break

        turns += 1
        reply = get_response(user_input)
        print(f"{Ai_assistant_Name}: {reply}")
```

```python
if __name__ == "__main__":
    main()
```
